```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

analisa_df = pd.read_csv('Penyebab Kematian di Indonesia yang Dilaporkan - Clean.csv')
analisa_df
```

```python
# 1. Total kematian pertahun dan penyebab penyakit
# Memanggil kolom dari Total Deaths, Year dan Cause
no1_df = analisa_df[['Total Deaths', 'Year', 'Type']]
no1_df
```

```python
# Menjumlahkan total kematian pertahun 
no1_df.groupby('Year')['Total Deaths'].agg(['sum', 'max'])
```

```python
# Menjumlahkan total kematian pertahun 
no1_df.groupby('Type')['Total Deaths'].agg(['sum'])
```

```python
# 2. Penyebab kematian dari non bencana alam dan penyakit
no2_df = analisa_df[['Cause', 'Type', 'Total Deaths']]
no2_df
```

```python
kr1 = no2_df['Type']=='Bencana Non Alam dan Penyakit'
no2_df[kr1]
```

```python
# 3. Penyebab kematian tertinggi dari bencana alam
no3_df = analisa_df[['Type', 'Total Deaths', 'Cause']]
no3_df
```

```python
no3_df.sort_values('Total Deaths', ascending=False).head(10)
```
